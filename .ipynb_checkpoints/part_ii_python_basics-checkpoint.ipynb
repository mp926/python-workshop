### Functions in Python

As in most languages, a function can be defined as a block of code that accepts some input, processes them in a certain way and gives us an out.  
Or it can be defined as block of code that performs a certain specific task

```python
# Defining a function in Python
def function_name(variable):
    return variable
    
# Calling the function
function_name('Learn Python')
```

```python
# Defining a more complex function
# This function will take in a list of names (string variables)
# and then return the number of characters in each name
def count_length_of_names(list_names):
    # declare an empty list that will hold the
    # length of each name
    length_list = []
    # Loop through each name
    for name in list_names:
        # Then using the built-in 'len' function get the length of the name
        length_list.append(len(name))
    # return the 'length_list'
    return length_list
```

```python
my_list = ['Kumar Sangakkara', 'Ricky Ponting', 'Sachin Tendulkar', 'Damien Martyn', 'VVS Laxman']

count_length_of_names(my_list)
```

## The SciPy ecosystem

The SciPy ecosystem is a collection of Python libraries that allows for easy scientific computation.  
The major libraries that makes up the ecosystem are  
1. Numpy - For heavy duty numerical computation, matrix manipulations
2. Scipy - A collection of a large number of scientific algorithms
3. Matplotlib - A library that allows for scientific plotting
4. Pandas - A library that allows for dataframe construction and manipulation
5. Jupyter - Which is the interactive notebook that you are currently working in

Since these libraries are not part of the core Python system, they must be installed separately. Which is why the Anaconda python package is so convenient. All these libraries are installed alongside base Python.  
So let's get into it. We'll look at numpy arrays, draw some plots, and then work on dataframes

### Numpy

```python
# First let's import the numpy library
# It's standard to import it as 'np'
import numpy as np

# Let's make a 1-D array
array_1 = np.array([1,2,3,4,5])
# We can manipulate the array in any way we want
array_2 = array_1 * 5
print(array_2)
```

```python
# Let's make a 2-D array
array_3 = np.array([[1,2],[3,4]])
print(array_3)
# It's basically a 2x2 matrix
# Numpy is a vast topic that can't be covered in an hour. Or even a day
# But it's the bedrock upon which scientific computing in Python is built on
```

### Matplotlib - Plotting with Python

A nice piece of history. The author of Matplotlib, John Hunter, created this library because back in the early 2000's, Python couldn't match up to the plotting capabilities of R. Some of the syntax and ideas in matplotlib are influenced by R design

```python
%matplotlib inline
```

```python
# Let's import matplotlib
# Just like importing numpy, this is the standard way to do it
import matplotlib.pyplot as plt

# Make an array.
# And here's a new numpy function 'arange'. Similar to range in base Python
array_x = np.arange(0.5, 25.5, 0.75)
# Apply a function to the array (Square it)
array_y = array_x **2

print(array_x)
print(array_y)
```

```python
# Plot the function
plt.plot(array_x, array_y)
```

```python
# Creating subplots

array_z = array_x **4

# 1st subplot
plt.subplot(211)
# What does '211' mean?
# Basically that there are 2 subplots and this one is in Row 1, Column 1
plt.plot(array_x, array_y)

# 2nd subplot
plt.subplot(212)
# This one is in Row 2, Column 1
plt.plot(array_x, array_z)
```

```python
# Creating a sin and cosine plot

array_l = np.arange(-5,5,0.1)
# Using 'sin' function in numpy library
array_sin = np.sin(array_l)
# Similar function for 'cosine'
array_cos = np.cos(array_l)

# 1st subplot
plt.subplot(211)
plt.plot(array_l, array_sin)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sin')

# 2nd subplot
plt.subplot(212)
plt.plot(array_l, array_cos)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Cos')

# To automatically set padding and optimally arrange the outlines
plt.tight_layout()
```

### Pandas - Working with datasets in Python

Pandas is a cool library built by Wes McKinney to allow for convenient data analaysis in Python. It's closely built after the **data.frame** structure in R

```python
# Importing the library
import pandas as pd
```

For the dataset we'll use some of the sampling data from my undergraduate thesis. These are blue swimming crabs caught by fishermen and their morphometrics

```python
# Reading a dataset
# Pandas has builtin functions that allow for reading different data formats
# Reading a CSV
all_data = pd.read_csv('data/bsb_catch_data.csv')
```

```python
all_data
```

```python
# Let's filter out the males
# For this we use something called a 'mask' which is basically the filter that we apply on this dataset

male_data=all_data[all_data['sex']=='M'].reset_index(drop=True)
```

```python
plt.figure(figsize=(7,10))

plt.scatter(male_data['carapace_width'],male_data['weight'],facecolors='white',edgecolors='black',s=30)#facecolors='#ffa64d'
plt.xlabel('CW\n(mm)')
plt.ylabel('W\n(g)')
x_values=plt.xlim()
y_values=plt.ylim()
plt.gca().yaxis.grid(False)
plt.gca().xaxis.grid(False)
plt.title('CW vs. W in males',fontsize=14)
```
